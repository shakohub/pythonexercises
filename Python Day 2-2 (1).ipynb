```python
#Fourth thing to teach (elif and else statements/conditionals)

#Well now we know about if but what if I want a condition dependent on another condition so a check within a check?
#Then we use elifs and else to make sure everything is taken care of
#Elif is = else if. So it will only be checked if the previous if or else if statement was false.
#An else statement is to handle anything else outside of the statements that you checked from above.
#So it's like a last resort or if all else fails then do this
#Look at if/elif they have conditions but else does not!
#remember indentation right after your colon.

#assigning some variable with numbers
people = 30
cars = 40
trucks = 15

#now do the same as before! And print the boolean value above
#True
if cars > people:
    print("We should take the cars.")
#doesn't get checked is skipped since the if statement was true
elif cars < people:
    print("We should not take the cars.")
#doesn't get checked is skipped since the if statement was true
else:
    print("We can't decide.")
    
#False
if trucks > cars:
    print("That's too many trucks.")
#True
elif trucks < cars:
    print("Maybe we could take the trucks.")
#Doesn't get checked is skipped since the elif statement was true
else:
    print("We still can't decide.")

#True
if people > trucks:
    print("Alright, let's just take the trucks.")
else:
    print("Fine, let's stay home then.")
    
#Before running the cell check which ones will print then
#check it against the cell once you run it
```

```python
# Seventh thing to teach (functions with loops now)

#lets try to do modulus % without using modulus!
#What do I need?
#Well I first need to think which kind of loop do I want to use?
#How do I determine that? Well do I have an exact amount times that this will occur always?
#If yes then i'll use a for loop, but if not then I should use a condition to check if I still need to do a loop or not

#I don't know how many times i'll need to subtract the second number from the first number so I use a while loop
#Next I have to think what is my condition going to be?
#When will I know that I have to stop the program?
#Ah the solution is I have to stop it when I reach anything that becomes negative
#Next I need to think about the amount of arguments I will need. Well I'm just subtracting one number from another
#So I need two arguments

#assignment is =
#testing is ==
def modulus_hard(arg1, arg2):
    #to check if you subtract number 2 from number 1 and we are still greater than zero then I will keep going
    while (arg1 - arg2) > 0:
        #take out number 2 from number 1 and reassign new value to arg1
        #Long form is arg1 = arg1 - arg2
        arg1 -= arg2
    #once we finish we have the remainderas arg1 so we return it
    return arg1

#first number give it is 3 and second number is 2
#3 - 2  > 0  ->  1 > 0 true
#new number 1 is going to be 3 - 2
# arg1 is now 1
#1 - 2 > 0  -> -1 > 0 false


#if you make an infinite loop
#you can use a keyword called break

#did I call the function?


#now I did
#same as saying 100 % 11
print(modulus_hard(100,11))
```

```python
# Eighth thing to teach (lists)

#Creating list to use that information
#Lists can hold multiple different types of data types or one type of data type
#lists are useful when you have a ton of information and need to store it somewhere.
#lists can be used in categories and for storing a large set of numbers.

#list of numbers
the_count = [1 ,2, 3, 4, 5,10]
#list of strings
fruits = ['apples', 'oranges' ,'pears' ,'apricots']
#list of numbers and strings
change = [1, 'pennies', 2, 'dimes', 3, 'quarters']

           #index  0    1    2    3    4     5    6
#pill container [| 1  | 2  |  3 | 4  | 5  | 10 |]

#normal in range for loop
for number in range(0,6):
    print(the_count[number])

# this first kind of for-loop goes through a list
# we call this an enhanced for-loop
for number in the_count:
    print(f"This is count {number}")

# same as above 
for fruit in fruits:
    print(f"A fruit of type: {fruit}")

# also we can go through mixed lists too
# notice we have to use {} since we don't know what's in it
for i in change:
    print(f"I got {i}")

# we can also build lists, first start with an empty one
elements = []

# then use the range function to do 0 to 5 counts
for i in range(0, 6):
    print(f"Adding {i} to the list.")
    # append is a function that lists understand
    # why is there a dot?
    elements.append(i)
#first iteration elements = [0]
#second iteration elements = [0,1]
#...
#sitxh iteration elements = [0,1,2,3,4,5]

# now we can print them out too
for element in elements:
    print(f"Element was: {element}")
    
# sort function ascending order  from 0 - 100 
# but when we have optinal variable caled reverse, you can set it to be True to go from descending order.  100 - 0
elements.sort(reverse = True)
print("Reversed sorted")
print(elements)

elements.sort()
print("Sorted Normally")
print(elements)
```
